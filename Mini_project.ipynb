```python
import os
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from skimage.io import imread
from skimage.transform import resize
```

```python
## creating empty sets 
Target = []
Images = []
Flat_data = []
```

```python
## Directories
Directory = '/content/sports'
Categories = ['Cricket ball' , 'Rugby ball leather' , 'Soccer ball' , 'Tennis ball']
```

```python
for category in Categories:
  ## Label endoing the value
  Index = Categories.index(category)

  ## path of the category
  path = os.path.join(Directory , category)

  for img in os.listdir(path):
    img_array = imread(os.path.join(path , img))
    
    ## resize function normalize the data as well

    img_resized = resize(img_array , (150,150,3))
    Flat_data.append(img_resized.flatten())
    Images.append(img_resized)
    Target.append(Index) 
```

```python
Flat_data = np.array(Flat_data)
Target = np.array(Target)
Images = np.array(Images)
```

```python
## Train test split
from sklearn.model_selection import train_test_split
```

```python
x_train ,x_test ,y_train ,y_test = train_test_split(Flat_data , Target , test_size = 0.2 , random_state = 0)
```

```python
x_train[0]
```

```python
y_train[0]
```

```python
## KNN MODEL
from sklearn.neighbors import KNeighborsClassifier
```

```python
Model_K = KNeighborsClassifier(n_neighbors=3)
Model_K.fit(x_train , y_train)
```

```python
from sklearn.metrics import accuracy_score , confusion_matrix
```

```python
accuracy_score(y_test , Model_K.predict(x_test))
```

```python
confusion_matrix(y_test , Model_K.predict(x_test))
```

```python
## GRIDSEARCHCV

from sklearn.model_selection import GridSearchCV
```

```python
param = {'n_neighbors':[3,4,5,6,8,10] }
model = KNeighborsClassifier()

model_1 = GridSearchCV(model , param , scoring='accuracy' , n_jobs=-1 , cv=4 , verbose =1)

model_1.fit(x_train , y_train)
```

```python
model_1.best_params_
```

```python
Model_K_GRID = model_1.best_estimator_
```

```python
accuracy_score(y_test , Model_K_GRID.predict(x_test))
```
